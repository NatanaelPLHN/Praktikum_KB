##### 3. Tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka. Masukkan informasi tersebut dalam dictionary dengan key sesuai nama informasinya. Tampilkan isi dari dictionary tersebut sebagus mungkin.

```python
dictionary={
    "Jumlah record"  : len(df),
    "Jumlah atribut" : len(df.columns),
    "Nama atribut"   : df.columns.tolist(),
    "Jumlah atribut dengan tipe angka" : len(df.select_dtypes(include=[int,float]).columns),
}

for key,value in dictionary.items():
    print(f"{key} : {value}")
```

###### Disini kita menggunakan len(df) untuk mencari tahu jumlah record yang ada di dalam dataset 
###### Disini kita menggunakan len(df.columns) untuk mencari tahu jumlah atribut yang ada di dalam dataset
######
###### Disini kita menggunakan len(df.select_dtypes(include=[int,float]).columns, untuk mencari jumlah record dari dataset yang memiliki tipe angka

```python
df.loc[29:, "Floor":]
```
