### Import library yang dipakai

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Buka dataset

```python
dfHouse = pd.read_csv("House_Rent_Dataset.csv")
```

```python
dfHouse.head(100)
```

### Grafik batang "Tenant Preferred"
##### Grafik yang menunjukkan jumlah dari tipe penghuni(Tenant) yang diharapkan menghuni properti.

```python
# menghitung jumlah data dari column "Tenant Preferred"
tenantPreferred_valCount = dfHouse['Tenant Preferred'].value_counts()

plt.figure(figsize=(20,10))

#membuat grafik batang
plt.bar(tenantPreferred_valCount.index, tenantPreferred_valCount)

# menampilkan jumlah data di atas tabel
for i,j in enumerate(tenantPreferred_valCount):
    plt.text(i,j, str(j), ha="center", va="bottom")

#label dan title
plt.xlabel("Tenant Preferred")
plt.ylabel("Jumlah")
plt.title("Kategori Tenant yang diharapkan menghuni properti")
plt.show()
```

### Grafik batang "Furnishing Status"
##### Grafik yang menunjukkan kondisi furnishing dari properti yang dijual.

```python
# menghitung jumlah data dari column "Furnishing Status"
furnishingStatus_valCount = dfHouse['Furnishing Status'].value_counts()

plt.figure(figsize=(20,10))

# membuat grafik batang
plt.bar(furnishingStatus_valCount.index, furnishingStatus_valCount, color=['red'])

# menampilkan jumlah data di atas tabel
for i,j in enumerate(furnishingStatus_valCount):
    plt.text(i,j, str(j), ha="center", va="bottom")

#label dan title
plt.xlabel("Furnishing Status")
plt.ylabel("Jumlah")
plt.title("Kondisi Furnishing Properti")
plt.show()
```

### Grafik batang "City"
##### Grafik yang menunjukkan jumlah properti yang disewakan di masing-masing kota.

```python
# menghitung jumlah data dari column "City"
city_valCount = dfHouse['City'].value_counts()

plt.figure(figsize=(20,10))

#membuat grafik batang
plt.bar(city_valCount.index, city_valCount, color=['green'])

# menampilkan jumlah data di atas tabel
for i,j in enumerate(city_valCount):
    plt.text(i,j, str(j), ha="center", va="bottom")

#label dan title
plt.xlabel("City")
plt.ylabel("Jumlah")
plt.title("Jumlah properti pada setiap City")
plt.show()
```

### Histogram "Bathroom"
##### Histogram yang menampilkan jumlah kamar mandi pada properti.


```python
plt.figure(figsize=(20,10))
plt.hist(dfHouse["Bathroom"], bins = 12)

#label dan title
plt.xlabel("Bathroom amount")
plt.ylabel("Jumlah")
plt.title("Jumlah kamar mandi pada properti")
plt.show()
```

### Count Plot
##### CountPlot yang menampilkan tipe area yang ada di properti pada masing-masing kota.

```python
#buat scatterplot
plt.figure(figsize=(20,10))
sns.countplot(data=dfHouse, x="Area Type", hue="City")

#title
plt.title("Tipe area di properti per kota")
plt.show()
```

### Box Plot BHK
##### Box plot yang menampilkan jumlah Kamar, Lorong, dan Dapur(Bedroom, Hall, Kitchen) pada setiap kota.

```python
plt.figure(figsize=(20,10))
sns.boxenplot(data = dfHouse, x='BHK', y='Size', hue="City")

#title
plt.title("Jumlah BHK pada properti setiap kota")
plt.show()
```

### Pie Chart
##### Pie chart yang menggambarkan persentase jumlah data "City"

```python
plt.figure(figsize=(16,30))

#baut pie chart
plt.pie(city_valCount, autopct= "%1.1f%%", explode= [0,0,0,0,0,0.1], labels=city_valCount.index)

#title dan label
plt.title("Persentase properti setiap City")
plt.legend(labels = city_valCount.index ,loc = "best")
plt.show()
```

### ScatterPlot Rent & Size
##### Scatterplot yang menggambarkan relasi antara harga sewa dan ukuran properti dengan hue kota.
###### Note : Di grafik ini kita dapat melihat beberapa outliers dari data yng kita punya.

```python
plt.figure(figsize=(25,8))

#buat scatterplot
sns.scatterplot(data=dfHouse, x="Rent", y="Size", hue="City")

#title
plt.title("Scatter Plot relasi harga sewa dan ukuran properti")
plt.show()
```

### Heatmap
##### Heatmap yang menampilkan korelasi antara beberapa atribut yang ada di dalam data

```python
plt.figure(figsize=(20,10))

sns.heatmap(dfHouse.corr(numeric_only=True), annot=True)

plt.title("Heatmap korelasi antara atribut")
plt.show()
```
